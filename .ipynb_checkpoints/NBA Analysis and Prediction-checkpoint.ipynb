```python
df['Offensive_score'] = df['PTS'] +df['Field_Goal_Percentage']+ df['3P%']+df['Assist']+df['Steals']
df['Defensive_score'] = df['Defensive_Rebounds'] +df['Blocks']+ df['Steals']
```

```python
fig,axes =plt.subplots(nrows=1,ncols=2,figsize=(15,5))
sns.barplot(ax = axes[0],x='POS',y='Offensive_score',data = df,palette = 'rocket');
sns.barplot(ax = axes[1],x='POS',y='Defensive_score',data = df,palette = 'rocket');
```

```python
best_offensive_player = df.groupby(['Player', 'Team']).agg({'Offensive_score': 'max'}).sort_values(by='Offensive_score', ascending=False).head(10)
best_offensive_player.plot(kind='barh', color='blue', figsize=(10, 5))
```

```python
best_defensive_player = df.groupby(['Player','Team']).agg({'Defensive_score':'max'}).sort_values(by='Defensive_score',ascending=False).head(10)
best_defensive_player.plot(kind='barh', color='black',figsize=(10,5))
```

Dropping columns which are highly correlated

```python
df1 = df.drop(columns = ['Player','POS','Team','Field_Goals_Made','Field_Goals_Attempted','3PM','3PA','3P%','Fantasy_Points','Double_Doubles','Triple_Doubles','Offensive_score','Defensive_score'],axis=1)
```

```python
df1.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = df1.drop(['PTS'],axis=1)
y = df1['PTS']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

Using Random Forest Regressor

```python
rfr = RandomForestRegressor(n_estimators=80,max_depth=5)
rfr.fit(X_train,y_train)
```

```python
y_pred = rfr.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```

```python
important_features_rfr = pd.Series(rfr.feature_importances_,index=X.columns)
important_features_rfr.nlargest(5).plot(kind = 'barh',figsize=(10,5))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

Using Linear Regression

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

Using KNN

```python
sc = StandardScaler()

X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=10)

knn.fit(X_train_scaled,y_train)
```

```python
y_pred_knn = knn.predict(X_test_scaled)
```

```python
r2_score(y_test,y_pred_knn)
```

# In terms of models, Linear Regression has the highest score of 0.966, followed by Random forest at 0.95 and knn at 0.74.

The most important features are: Minutes_Played,FTM,Turnover,FTA and Assist
